```python
import pandas as pd
import numpy as np
```

```python
path = 'C:/Users/jorge/OneDrive/Escritorio/ProyectoJacobo/EnvipeLimpia/TPer_vic2_Clean2017.csv'
df = pd.read_csv(path, low_memory=False)
df.head()
```

```python
n = df.columns.values.tolist()
m = round(df.shape[0]*.50)
l = []
for i in n:
    if df[i].value_counts().get("TTT") != None:
        if m < df[i].value_counts().get("TTT"):
            l.append(i)
            df = df.drop(i, 1)
```

```python
n = df.columns.values.tolist()
m = round(df.shape[0]*.50)
l = []
for i in n:
    if df[i].value_counts().get(99) != None:
        if m < df[i].value_counts().get(99):
            l.append(i)
            df = df.drop(i, 1)
```

```python
df.head()
```

```python
n = df.columns.values.tolist()
for i in n:
    if df[i].value_counts().get("TTT") != None:
        if df[i].mode().tolist()[0] == 'TTT':
            m = df.drop(df[(df[i] == 'TTT')].index)
            m= m[i].mode().tolist()
            df[i].replace({'TTT':m[0]}, inplace=True)
        else:
            df[i].replace({'TTT':df[i].mode().tolist()[0]}, inplace=True)
df.head()
```

```python
n = df.columns.values.tolist()
for i in n:
    if df[i].value_counts().get(99) != None:
        if df[i].value_counts().get(9) == None:
            df[i].replace({99 : 9},inplace = True)
df.head()
```

```python
df.to_csv("TPer_vic2_Clean2017_V3.csv")
```
